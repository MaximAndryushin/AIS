```python
from matplotlib import pyplot as plt
import numpy as np
```

## Гиперболическое уравнение регрессии: y = a + b / x

```python
def goal_func(x) :
    return 9.45 + -10.7 / x
def func(x) :
    return 9.45004 + -10.7 / x
x = np.linspace(1,100,1 << 20)
y1 = goal_func(x)
y2 = func(x)

plt.subplots(figsize=(10,6))
plt.plot(x, y1, label='Исходная')
plt.plot(x, y2, label='Найденная')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Гиперболическое уравнение регрессии')
plt.grid()
```

## Степенное уравнение регрессии: y = a * x ^ b

```python
def goal_func(x) :
    return -0.7 * x ** 2.4
def func(x) :
    return -1.72239 * x ** 2.19537
x = np.linspace(1,100,1 << 20)
y1 = goal_func(x)
y2 = func(x)

plt.subplots(figsize=(10,6))
plt.plot(x, y1, label='Исходная')
plt.plot(x, y2, label='Найденная')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Степенное уравнение регрессии')
plt.grid()
```

## Экспоненциальное уравнение регрессии: y = a * e ^ (b * x)


```python
def goal_func(x) :
    return 1.2 * np.exp(-3.1 * x)
def func(x) :
    return 1.24187 * np.exp(-3.18037 * x)
x = np.linspace(0,100,1 << 20)
y1 = goal_func(x)
y2 = func(x)

plt.subplots(figsize=(10,6))
plt.plot(x, y1, label='Исходная')
plt.plot(x, y2, label='Найденная')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Экспоненциальное уравнение регрессии')
plt.grid()
```
